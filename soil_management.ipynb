```python
import mysql.connector
import random
import tkinter as tk
from tkinter import ttk, messagebox
from faker import Faker
from datetime import datetime, timedelta
```

```python
fake = Faker()
```

```python
DB_CONFIG={
    "host":"localhost",
    "user":"root",
    "password":"2005",
    "database":"soil_management"
}
```

```python
conn = mysql.connector.connect(**DB_CONFIG)
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS Soil (
    id INT AUTO_INCREMENT PRIMARY KEY,
    farm_id INT NOT NULL,
    soil_type VARCHAR(255) NOT NULL,
    soil_ph VARCHAR(255) NOT NULL,  -- Ensure this column is present
    soil_moisture VARCHAR(255) NOT NULL,
    irrigation_method VARCHAR(255) NOT NULL,
    fertilizer_application_date DATE,
    fertilizer_type VARCHAR(255),
    yield_prediction INT
)
""")
conn.commit()
```

```python
#List
Soil_Type = ["Sandy", "Clay", "Silt", "Peat", "Chalk", "Loam"]
Soil_Ph = ["Acidic", "Neutral", "Alkaline"]
Soil_Moisture = ["Dry", "Moist", "Wet"]
Irrigation = ["Drip", "Sprinkler", "Flood", "Furrow", "Center Pivot"]
Fertilizer = ["Nitrogen", "Phosphorus", "Potassium", "Calcium", "Magnesium", "Sulfur", "Iron", "Manganese", "Copper", "Zinc", "Boron", "Molybdenum", "Chlorine"]
```

```python
def generate_soil_data():
    soil_type = random.choice(Soil_Type)
    soil_ph = random.choice(Soil_Ph)
    soil_moisture = random.choice(Soil_Moisture)
    irrigation_method = random.choice(Irrigation)
    fertilizer_type = random.choice(Fertilizer)
    fertilizer_application_date = fake.date_between(start_date="-2y", end_date="today")  # Application within last 2 years
    yield_prediction = random.randint(500, 5000)  # Yield in kg

    # Randomly generate a farm ID for association (You can change this logic)
    farm_id = random.randint(1, 100)  # Example: Random farm ID between 1 and 100

    return (farm_id, soil_type, soil_ph, soil_moisture, irrigation_method, fertilizer_application_date, fertilizer_type, yield_prediction)
```

```python
# Batch insert records for soil data in chunks
batch_size = 100000  # Insert 100,000 at a time for efficiency
total_records = 20000000  # Total records to insert

for i in range(0, total_records, batch_size):
    data_batch = [generate_soil_data() for _ in range(batch_size)]  # Generate random soil data
    
    # Execute batch insert for soil table
    cursor.executemany("""
        INSERT INTO soil (farm_id, soil_type, soil_ph, soil_moisture, irrigation_method, fertilizer_application_date, fertilizer_type, yield_prediction)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """, data_batch)
    conn.commit()
    
    print(f"{i + batch_size} records inserted for soil...")

print("✅ Soil data insertion completed successfully!")

# Close the database connection
cursor.close()
conn.close()
```
